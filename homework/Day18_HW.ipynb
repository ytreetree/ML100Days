```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np

#畫出完整的 cos 圖形

x = np.arange(0,3 * np.pi, 0.1)
y = np.cos(x)
plt.plot(x,y)
plt.savefig("HW18.png",dpi=300,format="png")
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

tan = X / Y
deg = np.arctan(tan)
plt.scatter(X, Y, c=deg)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
